# Line Charts 
A line chart or line plot or line graph or curve chart is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields (Wikipedia) 

````python
#import and read data
import pandas as pd
import matplotlib.pyplot as plt
 

df = pd.read_csv("./data_files/Worldbank GDP Data.csv", index_col = 0)
df = df.fillna(0)/1000000000
df
````

### Create a simple line plot

In order to create a simple line plot, we use ```pyplot.plot() ``` function. It is enough to supply just the y-values (GDP values) to the plot function. If only y-values are given, pyplot uses indexes as x-values. Optionally, x-values can also be supplied. When we call```pyplot.show() ```, the figure will be rendered and shown. Let's create a line plot of Turkey's GDP change over the years:

### Adding a Legend to the figure

Legends can be added to a figure by using ```pyplot.legend()``` function. This function takes a list strings as a parameter. The list should contain the name of each series in the same order they are added to the figure. 


````python
#increase the resolution of the figure
plt.figure(dpi=150)
#plt.figure(figsize=[10 , 5])
#plt.figure(dpi=150, figsize=[10 , 5])
 
#add 3 series of data to the line plot
plt.plot(df["Turkey"], label = "Turkiye") # alternatively: label = "Turkiye"
plt.plot(df["Switzerland"], label = "İsviçre") # alternatively: label = "İsviçre"
plt.plot(df["Korea"], label = "Güney Kore") # alternatively: label = "Güney Kore"

#add legends
#plt.legend(["Turkey", "Switzerland", "Korea" , "Albania"])
plt.legend() # label verildiğinde 

#you won't get any error if you add more (or less) legends than number of data series
#plt.legend(["Turkey", "Switzerland", "Korea", "Albania"])

plt.show()
````

### Change legend position 

Matplotlib, by default, tries to choose the "best" location for the figure legend. However, if you don't like the position, you can change the position by using *loc* parameter. The location codes are as follows: 
![Legend Location Codes](./images/legend_location.png)

You can use either string value or the location code. 

````python
#increase the resolution of the figure
plt.figure(dpi=150)

#add 3 series of data to the line plot
plt.plot(df["Turkey"])
plt.plot(df["Switzerland"])
plt.plot(df["Korea"])

#add legends
plt.legend(["Turkey", "Switzerland","Korea"], loc= 'upper right', framealpha = 0.3)
#plt.legend(["Turkey", "Switzerland","Korea"], loc= 6)

plt.show()
````

### Placing legend outside of the plot 

This requires a little bit more effort and understanding how matplotlib places legend on the plot. Matplotlib takes the point you specify as location (i.e upper right) of both plot and the legend box and places both on top of each other with some padding. For example, when you choose upper right as the *loc*, upper left corner of the plot and legend is brought together and then the legend is pushed a little inwards (padding).

The locations on the plot (anchor points) is held as proportions with respect to axes. For example: 

upper left = (0, 1)

lower center = (0.5 , 0)

upper right = (1 , 1)

If we want the legend outside of the figure, we need to explicitly tell matplotlib anchor point as the proportions with respect to axes. We can pass this anchor point by using *bbox_to_anchor* parameter.  

![Legend Explanation](images/legend_position.png)

````python
#increase the resolution of the figure
plt.figure(dpi=150)

#add 3 series of data to the line plot
plt.plot(df["Turkey"])
plt.plot(df["Switzerland"])
plt.plot(df["Korea"])

#add legends
plt.legend(["Turkey", "Switzerland","Korea"], loc= "lower center", bbox_to_anchor = (0.5,1.02), ncol = 3)

#add legends to the upper center outside of the plot. Increase the number of the columns to distribute legend horizontally
#bbox_to_anchor = (0.5, 1), ncol=3

#plt.legend(["Turkey", "Switzerland","Korea"], loc= "lower center", bbox_to_anchor = (0.5, 1.01),ncol = 3)

plt.show()
````

### Adding title, x-label and y-labels

These actions can be done by using the functions 

``` pyplot.title() ```

``` pyplot.xlabel() ```

``` pyplot.ylabel() ```

````python
#increase the resolution of the figure
plt.figure(dpi=150)

#add 3 series of data to the line plot
plt.plot(df["Turkey"])
plt.plot(df["Switzerland"])
plt.plot(df["Korea"])

#you won't get any error if you add more (or less) legends than number of data series
plt.legend(["Turkey", "Switzerland", "Korea"])

plt.title("Worldbank GDP Data", fontsize = 25, color = '#705e75') # r:red, b:blue, k:black, g:green, c:cyan
plt.xlabel("Calendar Years")
plt.ylabel("GDP (in billion US $)")

plt.show()
````

### Add markers and change linewidth, linestyle 

**marker styles:** https://matplotlib.org/api/markers_api.html#module-matplotlib.markers

**line styles:** {'-', '--', '-.', ':', ''}

**linewidth:** a float number

**other plot parameters:** https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

````python
#increase the resolution of the figure
plt.figure(dpi=150)

#add 3 series of data to the line plot
plt.plot(df["Turkey"], linestyle="--")
plt.plot(df["Switzerland"], marker = "o", markevery = slice(30,51,2)) # markevery = slice(30,51,2)
plt.plot(df["Korea"], linestyle = ":", linewidth = 4, color = "k")

#add legends
#plt.legend(["Turkey", "Switzerland","Korea"])

#you won't get any error if you add more (or less) legends than number of data series
plt.legend(["Turkey", "Switzerland", "Korea", "Albania"])

plt.title("Worldbank GDP Data")
plt.xlabel("Calendar Years")
plt.ylabel("GDP (in billion US $)")

plt.grid(b=True, axis = "y", linestyle = "--", color = "g")

plt.show()
````

### Adding text to the graph

We may want to add explanatory text into the figure. We can do this by using ``` pyplot.text() ``` function. We need to specify the coordinates with respect to values we see in the x and y axes as well as the text we want to include. 

Notice that GDP's dip around the year 2008 due to the economic crisis happened that year. Let's add an explanatory text to stress that: 

````python
#increase the resolution of the figure
plt.figure(dpi=150)

#add 3 series of data to the line plot
plt.plot(df["Turkey"])
plt.plot(df["Switzerland"])
plt.plot(df["Korea"])

# Uncomment next 3 lines if you need marker at 2008 for each country 
#plt.scatter(2008,df.loc[2008,"Turkey"])
#plt.scatter(2008,df.loc[2008,"Switzerland"])
#plt.scatter(2008,df.loc[2008,"Korea"])

#add legends
plt.legend(["Turkey", "Switzerland","Korea"])

plt.title("Worldbank GDP Data")
plt.xlabel("Calendar Years")
plt.ylabel("GDP (in billion US $)")

plt.text(2008,1300,"2008 \nCrisis",
        horizontalalignment = "center",
        fontsize = 8,
        rotation= 90 ,
        bbox = dict(boxstyle="larrow", facecolor = "g", alpha= 0.5)) # to add a bounding box try "bbox = dict(boxstyle="round")"

#plt.text(2008, df.loc[2008,"Korea"], round(df.loc[2008,"Korea"],2), color = "g")

plt.axvline(2008, linestyle = "--", color = "g", alpha = 0.3) # axhline(y_intercept) for horizontal lines

plt.axhline(1000)

#try following parameters as well:
#rotation= 'vertical',
#bbox = dict(boxstyle ='larrow', facecolor = 'g',alpha = 0.5)


plt.show()


````

Different box styles:  https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.BoxStyle.html#matplotlib.patches.BoxStyle

### Changing axis limits

Sometimes we may want to change the axis limits automatically created by Matplotlib. These limits can easily be altered by using ``` pyplot.xlim(lower,upper) ``` and ``` pyplot.ylim(lower,upper) ``` functions

````python
#increase the resolution of the figure
plt.figure(dpi=150)

#add 3 series of data to the line plot
plt.plot(df["Turkey"])
plt.plot(df["Switzerland"])
plt.plot(df["Korea"])

#add legends
plt.legend(["Turkey", "Switzerland","Korea"])


plt.title("Worldbank GDP Data")
plt.xlabel("Calendar Years")
plt.ylabel("GDP (in billion US $)")

plt.text(2008,1300,"2008 Crisis", horizontalalignment="center",
                                  verticalalignment="center",
                                  fontsize=8)

plt.xlim(1990, 2021)
plt.ylim(-100, 2100)

plt.minorticks_on()

plt.show()
````

### Adding Major and Minor Gridlines 

``` pyplot.grid(b=None, which='major', axis='both', **kwargs) ``` function is used for adding gridlines

**b:** *None* or True-False (Optional)

**which:** *'major'*, 'minor', 'both' (Optional)

**axis:** *'both'*, 'x', 'y' (Optional)

**kwargs:** Arguments regarding the style of gridlines. linewidth, linestyle, color etc.

P.S: We need to activate minorticks if we want the minor gridlines to appear. Use ``` pyplot.minorticks_on()```

````python

#increase the resolution of the figure
plt.figure(dpi=150)

#add 3 series of data to the line plot
plt.plot(df["Turkey"])
plt.plot(df["Switzerland"])
plt.plot(df["Korea"])

#add legends
#plt.legend(["Turkey", "Switzerland","Korea"])

#you won't get any error if you add more (or less) legends than number of data series
plt.legend(["Turkey", "Switzerland", "Korea", "Albania"])

plt.title("Worldbank GDP Data")
plt.xlabel("Calendar Years")
plt.ylabel("GDP (in billion US $)")

plt.text(2008,1250,"2008 Crisis", horizontalalignment="center",
                                  verticalalignment="center",
                                  fontsize=8)

#turn on minor ticks
plt.minorticks_on()

#add major gridlines:
#plt.grid(b=True)

#add major and minor gridlines
#plt.grid(b=True, which = 'both')

#major gridlines only on x-axis
#plt.grid(b=True, which = 'major', axis='y')

#major dashed gridlines in green
plt.grid(b=True, which = 'major', axis='both', linestyle= '--', color='g', alpha = 0.2)

plt.show()
````

### Modify Major Tick Marks

We can alter the places major ticks appear by using ``` pyplot.xticks() ``` and ``` pyplot.yticks() ``` functions. These functions take an array of major tick values as an argument.   

Currently x-axis show major ticks every 10 years and y- axis show tick marks every 250 billion. Let's change those to every 20 years and every 125 billion.  

````python
import numpy as np
#increase the resolution of the figure
plt.figure(dpi=150)

#add 3 series of data to the line plot
plt.plot(df["Turkey"])
plt.plot(df["Switzerland"])
plt.plot(df["Korea"])

#add legends
#plt.legend(["Turkey", "Switzerland","Korea"])

#you won't get any error if you add more (or less) legends than number of data series
plt.legend(["Turkey", "Switzerland", "Korea", "Albania"])

plt.title("Worldbank GDP Data")
plt.xlabel("Calendar Years")
plt.ylabel("GDP (in billion US $)")

plt.text(2008,1250,"2008 Crisis", horizontalalignment="center",
                                  verticalalignment="center",
                                  fontsize=8)
#Change x and y tick marks
#plt.xticks(np.arange(min(df.index),max(df.index)+3,20)) #[1960, 1980, 2000, 2020]

plt.xticks([1960, 1980, 2000, 2008, 2020])

plt.yticks(np.arange(0,max(df["Korea"]) + 200,500))

plt.savefig("my_figure.PNG", dpi = 450)

plt.show()
````
