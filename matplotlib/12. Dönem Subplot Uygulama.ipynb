```python
#import and read data
import pandas as pd
import matplotlib.pyplot as plt
 

df = pd.read_excel("./data_files/Worldbank GDP Data.xlsx", index_col = 0)
df = df.fillna(0)/1000000000
df
```

```python
empCount = [82, 91, 105, 112, 111, 109, 110, 110]
revenue= [0.6, 0.5, 0.7, 0.9, 0.6, 0.6, 0.8, 1]

quarters = [ "13Q1", "13Q2", "13Q3", "13Q4", "14Q1", "14Q2", "14Q3", "14Q4"]
```

```python
fig, ax = plt.subplots(2,2, dpi=150)

ax[0,0].plot(df["Turkey"])
ax[0,0].plot(df["Switzerland"])
ax[0,0].legend(["Turkey", "Switzerland"])
ax[0,0].set_xlabel("Years")
ax[0,0].set_ylabel("GDP")

ax[1,0].plot(df["Korea"])
ax[1,0].legend(["Korea"])
ax[1,0].set_xlabel("Years")
ax[1,0].set_ylabel("GDP")

ax[0,1].plot(quarters, empCount, marker="o")
ax[0,1].set_xlabel("Quarters")
ax[0,1].set_ylabel("Employee Count")


ax[1,1].bar(quarters, revenue)
ax[1,1].set_xlabel("Quarters")
ax[1,1].set_ylabel("Revenue")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,2, sharex='col', dpi=150)

ax[0,0].plot(df["Turkey"])
ax[0,0].plot(df["Switzerland"])
ax[0,0].legend(["Turkey", "Switzerland"])
ax[0,0].set_ylabel("GDP")

ax[1,0].plot(df["Korea"])
ax[1,0].legend(["Korea"])
ax[1,0].set_xlabel("Years")
ax[1,0].set_ylabel("GDP")

ax[0,1].plot(quarters, empCount, marker="o")
ax[0,1].set_ylabel("Employee Count")


ax[1,1].bar(quarters, revenue)
ax[1,1].set_xlabel("Quarters")
ax[1,1].set_ylabel("Revenue")
ax[1,1].set_xticklabels(quarters, Rotation= 90)

plt.tight_layout()
plt.show()
```
