# Subplots

In some cases, we may want to have multiple graphs in one figure. So far, in all figures we created, we only had one graph. That is why we didn't need to work with axes objects. When we have more than one plot in a single figure, we work with multiple axes objects in a single figure object. 

![Subplots](images/subplot.jpg)

In order to set up a space for each plot, we need to specify the number of plots and their layout at the beginning. We can do this by using ``` pyplot.subplot(nrows, ncols, index)``` where *nrows* specifies the number of rows, *ncols* speficies number of columns and *index* specifies where to put the next plot within the layout. ``` pyplot.subplot(nrows, ncols, index)``` function returns an axes object and we use that object to create our plots. Alternatively, these three parameters can be written one after another: ``` pyplot.subplot(221)``` -> nrows:2 , ncols:2 index:1.    

Now let's create three separate line graphs for Turkey, Switzerland and Korea in the same figure.


````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel("data_files\Worldbank GDP Data.xlsx",index_col=0)

#Fill missing values with 0 
df = df.fillna(0)

#express in billions of dollars
df = df /1000000000

df.tail()
````

````python
plt.figure(dpi=150)

#-----Subplot 1 Turkey-------# 
ax1 = plt.subplot(3,1,1)

ax1.plot(df["Turkey"])

#Add title and Labels
ax1.set_xlabel("Calendar Years")
ax1.set_ylabel("GDP ")
ax1.text(2008, df.loc[2008,"Turkey"], "2008 Economic Crisis",
        horizontalalignment='center',
        verticalalignment='center',
         fontsize=8)

#-----Subplot 2 Switzerland-------# 

ax2 = plt.subplot(3,1,2)

ax2.plot(df["Switzerland"])

#Add title and Labels
ax2.set_xlabel("Calendar Years")
ax2.set_ylabel("GDP")
ax2.text(2008, df.loc[2008,"Switzerland"], "2008 Economic Crisis",
        horizontalalignment='center',
        verticalalignment='center',
        fontsize=8)

#-----Subplot 3 Korea-------# 
ax3 = plt.subplot(3,1,3)

ax3.plot(df["Korea"])
#Add title and Labels
ax3.set_xlabel("Calendar Years")
ax3.set_ylabel("GDP")
ax3.text(2008, df.loc[2008,"Korea"], "2008 Economic Crisis",
        horizontalalignment='center',
        verticalalignment='center',
         fontsize=8)

plt.show()
````

### Adjust spacing

As you can see the plots above are too close to each other. Let's adjust the spacing by using ``` subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)``` function. By default these values are: 

``` python

left = 0.125  # the left side of the subplots of the figure
right = 0.9   # the right side of the subplots of the figure
bottom = 0.1  # the bottom of the subplots of the figure
top = 0.9     # the top of the subplots of the figure
wspace = 0.2  # the amount of width reserved for space between subplots,
              # expressed as a fraction of the average axis width
hspace = 0.2  # the amount of height reserved for space between subplots,
              # expressed as a fraction of the average axis height
    ```

````python
plt.figure(dpi=150)

#-----Subplot 1 Turkey-------# 
ax1 = plt.subplot(3,1,1)

ax1.plot(df.index.values, df["Turkey"], color= "b")

#Add title and Labels
ax1.set_xlabel("Calendar Years")
ax1.set_ylabel("GDP ")
ax1.text(2008, df.loc[2008,"Turkey"], "2008 Economic Crisis",
        horizontalalignment='center',
        verticalalignment='center',
         fontsize=8)
ax1.legend(["Turkey"])
#-----Subplot 2 Switzerland-------# 

ax2 = plt.subplot(3,1,2)

ax2.plot(df.index, df["Switzerland"], color= "g")

#Add title and Labels
ax2.set_xlabel("Calendar Years")
ax2.set_ylabel("GDP")
ax2.text(2008, df.loc[2008,"Switzerland"], "2008 Economic Crisis",
        horizontalalignment='center',
        verticalalignment='center',
        fontsize=8)
ax2.legend(["Switzerland"])
#-----Subplot 2 Korea-------# 
ax3 = plt.subplot(3,1,3)

ax3.plot(df.index, df["Korea"], color= "r")
#Add title and Labels
ax3.set_xlabel("Calendar Years")
ax3.set_ylabel("GDP")
ax3.text(2008, df.loc[2008,"Korea"], "2008 Economic Crisis",
        horizontalalignment='center',
        verticalalignment='center',
         fontsize=8)
ax3.legend(["Korea"])

plt.subplots_adjust(hspace = 1)

plt.show()
````

### More complex layouts

In order to achieve more complex plot layouts in the same figure, we need to use *gridspec*. When we add a gridspec in a Figure, it divides the figure area in grids and we can specify the grids we want our plot to appear. In order to, connect a gridspec to a figure, we can use ```Figure.add_gridspec(nrows, ncols)``` function. In order to use gridspec, we use ```Figure.add_subplot()``` function to create subplots

````python
#create a figure and get the handle
fig = plt.figure(dpi=150)

#add a gridspec size of 3 x 4
gs = fig.add_gridspec(3, 4)

#add first figure
ax1 = fig.add_subplot(gs[0:2, 0:2])
ax1.plot(df.index, df["Korea"], color= "r")
ax1.set_title('gs[0:2, 0:2]')

ax2 = fig.add_subplot(gs[1:3, 2:4])
ax2.plot(df.index, df["Korea"], color= "r")
ax2.set_title('gs[1:3, 2:4]')

ax3 = fig.add_subplot(gs[0, 2:4])
ax3.plot(df.index, df["Korea"], color= "r")
ax3.set_title('gs[0, 2:4]')

ax4 = fig.add_subplot(gs[2, 0:2])
ax4.plot(df.index, df["Korea"], color= "r")
ax4.set_title('gs[2, 0:2]')

plt.subplots_adjust(hspace = 0.7, wspace= 0.5)

plt.show()
````

### Shared X and/or Y axis 

At times we may want x or y axis to be shared among the graphs. In our example graphs we created above, x axis are common in all three and we may not need to put it three times. In order to achieve this, we need to use a slightly different approach and use a different function when creating the subplots. ``` pyplot.subplots(nrows, ncols)``` will create a figure object and as many axes objects as needed then returns us to use. Syntax: 

``` python
 fig, ax = plt.subplots(2,2) # since we created a 2x2 grid, ax is a 2x2 array of axes objects  
 ax[0,0].plot(....)
 ax[0,1].plot(....)
 
```  

In addition to nrows and ncols, we need to specify which axis are to be shared. We set *sharex* and/or *sharey* parameters to do that. Both can take on the following values: 

**True, False, 'all', 'row', 'col'**

For example, if *sharex='col'* then only the plots in the same column will share x - axis

````python

#-----Subplot 1 Turkey-------# 
#fig, ax = plt.subplots(3,1, sharex=True)
fig, ax = plt.subplots(3,1, sharex='all', sharey = 'all') # y-axis will have the same scale

ax[0].plot(df.index.values, df["Turkey"], color= "b")

#Add title and Labels
ax[0].set_ylabel("GDP ")
ax[0].text(2008, df.loc[2008,"Turkey"], "2008 Economic Crisis",
        horizontalalignment='center',
        verticalalignment='center',
         fontsize=8)
ax[0].legend(["Turkey"], fontsize=8)
#-----Subplot 2 Switzerland-------# 

ax[1].plot(df.index, df["Switzerland"], color= "g")

#Add title and Labels
ax[1].set_ylabel("GDP")
ax[1].text(2008, df.loc[2008,"Switzerland"], "2008 Economic Crisis",
        horizontalalignment='center',
        verticalalignment='center',
        fontsize=8)
ax[1].legend(["Switzerland"], fontsize=8)

#-----Subplot 2 Korea-------#
ax[2].plot(df.index, df["Korea"], color= "r")
#Add title and Labels
ax[2].set_xlabel("Calendar Years")
ax[2].set_ylabel("GDP")
ax[2].text(2008, df.loc[2008,"Korea"], "2008 Economic Crisis",
        horizontalalignment='center',
        verticalalignment='center',
         fontsize=8)
ax[2].legend(["Korea"], fontsize=8)

fig.set_dpi(150)

plt.subplots_adjust(hspace = 0.2)

plt.show()
````

### Second y-axis 

If we want to plot two related graphs together but the scales are different, we can use a secondary y axis. Let's recreate the example graph seen in the slides.  

````python
import matplotlib.ticker as ticker
empCount = [82, 91, 105, 112, 111, 109, 110, 110]
revenue= [0.6, 0.5, 0.7, 0.9, 0.6, 0.6, 0.8, 1]

quarters = [ "13Q1", "13Q2", "13Q3", "13Q4", "14Q1", "14Q2", "14Q3", "14Q4"]

#ax1 = plt.subplot(2,2,(1,2)): tek bir axes objesi
#ax1.scatter(....)

#fig, ax = plt.subplots(2,2): 4 adet axes objesi
#ax[0].scatter(...)

plt.figure(dpi=150)

ax1 = plt.subplot(1,1,1)
h1 = ax1.bar(quarters, revenue, label = "Quartery Revenue")
ax1.set_ylabel("Revenue (Millions)")
ax1.set_ylim(0,1.2)

#for adding $ sign to y-axis
formatter = ticker.FormatStrFormatter('$%1.1f')
ax1.yaxis.set_major_formatter(formatter)

#copy axes 1 to create axes2
ax2 = ax1.twinx()

h2 = ax2.plot(empCount , marker="o", label= "Size of Salesforce")
ax2.set_ylabel("Number of Sales Employees")
ax2.set_ylim(0,120)

plt.legend([h1, h2[0]],["Quartery Revenue", "Size of Salesforce"], loc = "lower center", bbox_to_anchor=(0.5,1.02), ncol = 2)

plt.show()
````

### Formatting the plot to get rid of the second y axis

In the slides, we discussed that secondary axis may not be desirable in some settings. Here is how you can remove it...

````python
import matplotlib.font_manager as fm
empCount = [82, 91, 105, 112, 111, 109, 110, 110]
revenue= [0.6, 0.5, 0.7, 0.9, 0.6, 0.6, 0.8, 1]

quarters = [ "13Q1", "13Q2", "13Q3", "13Q4", "14Q1", "14Q2", "14Q3", "14Q4"]

fig, ax1 = plt.subplots()
ax1.bar(quarters, revenue, color="#5173BE")
ax1.set_ylabel("Revenue (Millions)")
ax1.set_ylim(0,1.2)

# #write values on bar graph 
for i in range(len(revenue)):
    txt = "$" + str(revenue[i]) 
    ax1.text(i, revenue[i] , txt,
            horizontalalignment='center',
            verticalalignment='top',
            fontsize=8,
            color="w",
            fontproperties=fm.FontProperties(weight="semibold"))
    
ax1.text(0, 0.75 , "Revenue \n(Millions)",
            horizontalalignment='center',
            verticalalignment='top',
            fontsize=8,
            color="#85BCF7",
            fontproperties=fm.FontProperties(weight="semibold"))

# #don't show y-axis
ax1.get_yaxis().set_visible(False)
# ax1.set_frame_on(False)
ax1.spines["top"].set_visible(False)
ax1.spines["left"].set_visible(False)
ax1.spines["right"].set_visible(False)

# #copy axes 1 to create axes2
ax2 = ax1.twinx()


ax2.plot(empCount,marker='o',color="#1F4DB4")
ax2.set_ylabel("# of Sales Employees")
ax2.set_ylim(0,120)
ax2.set_label("Size of Salesforce")

for i in range(len(empCount)):
    ax2.text(i, empCount[i] + 2 , str(empCount[i]),
            horizontalalignment='center',
            verticalalignment='bottom',
            fontsize=8,
            color = '#1F4DB4',
            fontproperties=fm.FontProperties(weight="semibold"))
    
ax2.text(0, empCount[0] + 30 , "# of Sales \nEmployees",
            horizontalalignment='center',
            verticalalignment='top',
            fontsize=8,
            color="#1F4DB4",
            fontproperties=fm.FontProperties(weight="semibold"))

# #don't show y-axis
ax2.get_yaxis().set_visible(False)
# ax2.set_frame_on(False)
ax2.spines["top"].set_visible(False)
ax2.spines["left"].set_visible(False)
ax2.spines["right"].set_visible(False)

fig.set_dpi(150)

plt.show()
````
